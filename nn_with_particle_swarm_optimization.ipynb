```python
import torch
from torch.utils.data import Dataset,DataLoader
```

```python
import pandas as pd
data=pd.read_csv('Bank_Personal_Loan_Modelling.csv')
data.info()
```

```python
data.head()
```

```python
data.drop(columns=['ID','ZIP Code',],inplace=True)
```

```python
data.head()
```

```python
col=data.pop('Personal Loan')
data[col.name]=col.values
data.head()
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
features_train,features_test,labels_train,labels_test=train_test_split(data.iloc[:,:-1],data.iloc[:,-1],test_size=0.20)
```

```python
torch.tensor(features_train.values)
```

```python
class loan_dataset_implementation(Dataset):
    def __init__(self,feat,lab):
        self.features=torch.tensor(feat.values)
        self.labels=torch.tensor(lab.values)
    def __len__(self):
        return len(self.labels)
    def __getitem__(self,idx):
        return self.features[idx],self.labels[idx]
```

```python
loan_train_dataset=loan_dataset_implementation(features_train,labels_train)
loan_test_dataset=loan_dataset_implementation(features_test,labels_test)
wrapped_train_data=DataLoader(loan_train_dataset,batch_size=30,shuffle=True)
wrapped_test_data=DataLoader(loan_test_dataset,batch_size=30,shuffle=True)
```
